## Imputation

```python
imp = SimpleImputer(missing_values = np.nan, strategy = 'median') #"axis"has been moved from pararameter of SimpleImputer
median_imp_df = pd.DataFrame(imp.fit_transform(df))
median_imp_df
```

```python
plt.rcParams['figure.figsize'] = (20, 16)
sns.heatmap(median_imp_df.corr())
plt.show()
median_imp_df.corr()
```

```python
#variance inflation factor，VIF
vif = [variance_inflation_factor(median_imp_df.values, median_imp_df.columns.get_loc(i)) for i in median_imp_df.columns]

#當VIF<10,說明不存在多重共線性；當10<=VIF<100,存在較強的多重共線性，當VIF>=100,存在嚴重多重共線性
new_df = []
for idx, val in enumerate(vif):
    if (val <=3):
        print(f"Attr{idx}--Less collinearity==> {val}")
        new_df.append(median_imp_df.iloc[:, idx])
# Attr6
```

```python
#Drop out some features with collinearity and generate a new dataframe
vif_df = DataFrame(new_df)
vif_df = vif_df.transpose()
vif_df
```

## Spilt Training set & Testing set

```python
# vif_df(has been processing)
x = vif_df.iloc[:, 0: len(vif_df.columns)-1]
y = vif_df.iloc[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x
```

## Logistic Regression

```python
threshold = 1e-6
#cv = K-fold cross validation
clf1 = LogisticRegressionCV(penalty='l1', solver='liblinear', cv = 10) # lasso 
clf2 = LogisticRegressionCV(penalty = 'l2', cv = 10) #ridge
```

```python
model_list = (clf1, clf2)

for clf in model_list:
    clf.fit(x_train, y_train)
    feature_weight = clf.coef_
    # Predict values using the training data
    lr_cv_predict_train = clf.predict(x_train)
    print("Accuracy against training data: {0:.4f}".format(metrics.accuracy_score(y_train, lr_cv_predict_train)))
    print()
    # Predict values using the test data
    lr_cv_predict_test = clf.predict(x_test)
    print("Accuracy against test data: {0:.4f}".format(metrics.accuracy_score(y_test, lr_cv_predict_test)))
    print()
    print(f"the number of feature = {clf.n_features_in_}")
    print(feature_weight)
    print("The validation score of model", clf.score(x_test, y_test))
    print(f"The number of selected features = {np.sum(abs(feature_weight) > threshold)}")
   
```

## Extra Test (not useless)

```python
listC = 10.0**np.arange(-4,4) #give the range that we can search
parameter = {'C':listC}
lr = LogisticRegression(penalty = 'l1', solver='liblinear')
clf = GridSearchCV(lr, parameter)

clf.fit(x_train, y_train)
print("The best parameter is", clf.best_params_)
print("The best score is", clf.best_score_)
```

```python
# Using the best param to do it again
lr = LogisticRegression(penalty = 'l1', solver='liblinear', C = 0.01)
lr.fit(x_train, y_train)

print("Training score of Logistic Regression model is", lr.score(x_train, y_train))
print("Testing score of Logistic Regression model is", lr.score(x_test, y_test))
```
